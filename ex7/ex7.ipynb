# Image compression with kmeans

```python
K = 16 # number of colors after compression

A = mpl.image.imread('data/bird_small.png')
A /= 255 
X = A.reshape(-1, 3)
fit, idx = funcs.runkMeans(X,K)
centers = fit.cluster_centers_
X_recovered = centers[idx,:].reshape(A.shape)
```

```python
# Display the original image, rescale back by 255
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].imshow(A*255)
ax[0].set_title('Original')
ax[0].grid(False)

# Display compressed image, rescale back by 255
ax[1].imshow(X_recovered*255)
ax[1].set_title('Compressed, with %d colors' % K)
ax[1].grid(False)
```
